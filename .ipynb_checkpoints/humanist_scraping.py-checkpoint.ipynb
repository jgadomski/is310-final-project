```python
import requests
response = requests.get("https://humanist.kdl.kcl.ac.uk/")

print(response.status_code)
print(response.headers)
```

```python
soup = BeautifulSoup(response.text, features = "html.parser")
soup
```

```python
links = soup.find_all('a')
link_dict = {}
for link in links:
    if 'Volume' in link.text:
        link_text = link.get_text()
        link_href = link.get('href')
        link_dict[link_text] = link_href
    
        print(link_text,link_href)

link_dict    
```

```python
original = 'https://humanist.kdl.kcl.ac.uk'
original += link_dict['Volume 1 5/87-5/88']
original
```

```python
original2 = 'https://humanist.kdl.kcl.ac.uk'
original2 += link_dict['Volume 25/88-5/89']
original2
```

```python
def scrape_screenplay(url):
    response = requests.get(url)
    html_string = response.text
    return html_string
```

```python
scrape_screenplay(original2)
```
